```python
import json
import xmltodict
```

```python
# coco 파일로 파일 이름 불러오기
file = open('./Train/CoCo/coco_astrophysics.json')
js = json.load(file)
```

```python
# coco 파일 이름 저장하기
file_name = []
for i in js['images']:
  file_name.append(i['file_name'])
```

# XML TO JSON

```python
# xml -> json 코드
for j in range(137708): # 137708
  filexml = './Train/Pascal/Astrophysics/' + file_name[j][:-4] + '.xml'
  filejson = './Train/Pascal/json/' + file_name[j][:-4] + '.json'
  
  with open(filexml,'r', encoding='UTF-8') as f:
      xmlString = f.read()
  
  jsonString = json.dumps(xmltodict.parse(xmlString), indent=4)
  
  with open(filejson, 'w', encoding='UTF-8') as f:
      f.write(jsonString)
```

# 절대좌표 to 상대좌표 bbox

```python
# 절대좌표 -> 상대좌표 정사각형
def convert(size, box):
    dw = 1./size[0]
    dh = 1./size[1]
    x = (box[0] + box[1])/2.0
    y = (box[2] + box[3] + 840)/2.0
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return [x,y,w,h]

w = int(1920)
h = int(1920)

코드 셀 <3SoFppsUKrYD>
```
